```python
def get_image_shape(slice_path):
    # Open one image to get the shape (width, height)
    img = Image.open(slice_path).convert('RGB')
    return img.size  # Returns (width, height)
```

```python
import os

root = "/content/drive/MyDrive/MachineLearningProject/CrowdedObjectsCounting/HoneyBeesCounting/BeesDatasets"  # Change this to your target folder

```

```python
images = ['1283.h5', '1285.h5', '1288.h5', '1290.h5', '1292.h5', '1305.h5', '1307.h5', '1311.h5', '13194.h5', '1321.h5', '13217.h5', '1336.h5', '134.h5', '13404.h5', '13406.h5', '13446.h5', '1345.h5', '1347.h5', '135.h5', '1350.h5']
file_count = len(images)
print(f"Total files: {file_count}")
renamed_files = []
for file in images:
    # Extract the base filename (without extension) and add sequence
    base_name, ext = file.split('.')
    renamed_files.append(f"{base_name}_01_01.{ext}")
    renamed_files.append(f"{base_name}_01_02.{ext}")
    renamed_files.append(f"{base_name}_01_03.{ext}")
    renamed_files.append(f"{base_name}_02_01.{ext}")
    renamed_files.append(f"{base_name}_02_02.{ext}")
    renamed_files.append(f"{base_name}_02_03.{ext}")


print(len(renamed_files))
print(renamed_files)
```

```python
existing_files = []
non_existing_files = []

# Check if each renamed file exists
for file in renamed_files:
  # Check if the file exists
  new_path = root + file
  if os.path.exists(new_path):
      existing_files.append(new_path)
  else:
      non_existing_files.append(new_path)

# Print the results
print(f"Length: {len(existing_files)}")
print("Existing Files:", existing_files)
print(f"Length: {len(non_existing_files)}")
print("Non-Existing Files:", non_existing_files)

file_data = {
    "test_datasets": existing_files,
    "total": len(existing_files)
}
with open("/content/drive/MyDrive/MachineLearningProject/CrowdedObjectsCounting/HoneyBeesCounting/bees_val_paths.json", "w") as json_file:
    json.dump(file_data, json_file, indent=4)
```
