### For each label we have img_id in dataId

```python
s=0
f=0
for label in labells['labels']:
    img_name=label['dataId']
    img_name=img_name[:-4]
    labels=label['label']
    s+=1
#    print(img_name)
    if labels is not None:
        with open(os.path.join('All/ground-truth/',img_name+'.json'),'w') as ff:
            json.dump(labels,ff)
            f+=1
print('s',s)
print('g',f)
```
